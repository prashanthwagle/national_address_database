```python
import pandas as pd
import geopandas as gpd
import requests
import os
from tqdm import tqdm
```

```python
d = '../data/staging'
```

```python
dfs = []

for file in tqdm(os.listdir(d)):
    if file.split('.')[-1]!= 'xz':
        continue
    
    dfs.append(pd.read_csv(os.path.join(d, file), dtype={'geoid20':object}))

big_df = pd.concat(dfs)
big_df
```

```python
big_df=big_df[~big_df['geoid20'].isnull()] # remove nulls
big_df
```

```python
big_df['county'] = big_df['geoid20'].apply(lambda x: x[:5])
big_df
```

```python
export_dir = '../data/county_address'

for county in tqdm(big_df['county'].unique()):
    pdf = big_df[big_df['county'] == county]
    del pdf['county']
    pdf.to_csv(os.path.join(export_dir, county)+'.csv.xz', index=False)
```
